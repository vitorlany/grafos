```python
import networkx as nx
import urllib.request
```

```python
url = "https://raw.githubusercontent.com/vitorlany/grafos/main/graph-test-100.txt"
```

```python
graph = nx.Graph()

with urllib.request.urlopen(url) as f:
    for line in f:
        u, v = map(int, line.decode().strip().split())
        graph.add_edge(u, v)
```

```python
nx.draw(graph, with_labels=True)
```
